# OUSV Model by Schobel & Zhu (1999)

Schöbel R, Zhu J (1999) Stochastic Volatility With an Ornstein–Uhlenbeck Process: An Extension. Rev Financ 3:23–46. https://doi.org/10.1023/A:1009803506170

```python
pf.OusvFft.init_benchmark()
```

## Table 1A

```python
sheet_no = 5
m, p, rv = pf.OusvFft.init_benchmark(sheet_no)
m0, *_ = pf.OusvUncorrBallRoma1994.init_benchmark(sheet_no)
p
```

```python
rhos = [-1, -0.75, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1]

for rho in rhos:
    m.rho = rho
    print(np.round(m.price(**rv['args_pricing']), 3) - p[rho].values)
```

```python
### AVBS price
np.round(m0.price(**rv['args_pricing']), 3) - p['AVBS'].values
```
